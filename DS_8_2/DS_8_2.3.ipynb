```python
df_vis = my_df.query('price < 8').query('year == 2015').sort_values('price')
df_vis
plt.figure(figsize = (10, 6))
plt.plot(df_vis['price'], df_vis['reviews'],'*r', label = 'reviews/price rate')
plt.xlabel('Price')
plt.ylabel('Reviews', color = 'olive', fontsize = 'large')
plt.legend(loc = 'best')
plt.title('Analytics', color = 'red', fontsize = 'small')
plt.grid()
plt.show()
```

```python
my_df.groupby(['genre']).agg(['min', 'max'])['price']
```

Відповідь: Максимальна ціна для жанру Fiction:  - 82

Відповідь: Мінімальна ціна для жанру Fiction: 0

```python
df_autor = my_df[['author', 'name']].groupby(['author']).agg('count')
df_autor.sort_values('name')
```

Відповідь: Якої розмірності вийшла таблиця? Відповідь: - 248 авторов

Відповідь: Який автор має найбільше книг? Відповідь: - Jeff Kinney

Відповідь: Скільки книг цього автора? Відповідь: - 12

```python
df_rat = my_df[['author', 'user_rating']].groupby(['author']).agg('mean')
df_rat.sort_values('user_rating')
```

Відповідь: У якого автора середній рейтинг мінімальний? Відповідь: - Donna Tartt

Відповідь: Який у цього автора середній рейтинг? Відповідь: - 3.9

```python
df_com = pd.concat([df_autor, df_rat], axis=1)
df_com.sort_values(by=['name', 'user_rating'])
```
