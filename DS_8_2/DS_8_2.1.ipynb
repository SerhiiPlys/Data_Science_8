```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"
my_df = pd.read_html(url)
df_work = my_df[12]
#print(df_work)
```

```python
df_work = df_work.replace("—", np.nan)
print(df_work)
```

```python
df_work['2014'] = pd.to_numeric(df_work["2014"])
df_work['2019'] = pd.to_numeric(df_work["2019"])
print(df_work.dtypes)
```

Delete end row

```python
df_work = df_work.drop(27, axis=0)
print(df_work)
```

```python
val = {'1950': df_work['1950'].mean(), '1960':df_work['1960'].mean(), '1970':df_work['1970'].mean(), 
       '1990':df_work['1990'].mean(), '2000':df_work['2000'].mean(), '2012':df_work['2012'].mean(), 
       '2014':df_work['2014'].mean(), '2019':df_work['2019'].mean()}
df_work = df_work.fillna(value = val)
df_work
```

```python
s = df_work[df_work['2019'] > df_work['2019'].mean()]
#print(s)
print(s[['Регіон','2019']])
```

```python
print(df_work[df_work['2014'] == df_work['2014'].max()][['Регіон']])
```

```python
plt.barh(df_work['Регіон'],df_work['2019'])
plt.xlabel('Кількість', fontsize='small', color='black')
plt.ylabel('Регіон', fontsize=12, color='maroon')
plt.title('Народжуваність по регіонам 2019', fontsize=15)
plt.show()
```

```python
plt.barh(df_work['Регіон'],df_work.mean(axis=1, numeric_only=True))
plt.xlabel('Кількість', fontsize='small', color='black')
plt.ylabel('Регіон', fontsize=12, color='olive')
plt.title('Середня Народжуваність по регіонам 1950-2019', fontsize=15)
plt.show()
```

```python
plt.plot(df_work.columns.drop('Регіон'), df_work.mean(numeric_only=True))
plt.xlabel('Рік', fontsize='small', color='black')
plt.ylabel('Кількість', fontsize=12, color='maroon')
plt.title('Середня народжуваність по рокам 1950-2014', fontsize=15)
plt.show()
```
