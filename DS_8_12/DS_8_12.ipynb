```python
import spacy   #импортим библиотеку
nlp = spacy.load('en_core_web_sm') # импортим словарь для англ языка
# Токенізація
doc = nlp(input_txt)
tokens = [token.text for token in doc]
print(tokens)
# Удаление знаков препинания из списка токенов
tokens_without_punct = [token.text for token in doc if not token.is_punct]
print(tokens_without_punct)
# удаляем из списка токенов стопслова - то слова, которые обычно игнорируются при обработке текста в
# задачах обработки естественного языка (Natural Language Processing, NLP).
# Эти слова считаются неинформативными, так как они встречаются очень часто в текстах, но не несут особой смысловой нагрузки и мало влияют на смысл предложений.
tokens_without_stopword = [token.text for token in doc if not token.is_stop]
print(tokens_without_stopword)
```

```python
# получаем в конце концов список токенов чистый - лишь слова имеющие смысловую нагрузку
# Функция для фильтрации токенов
def is_token_valid(token):
    return not token.is_punct and not token.is_stop

# Получение списка токенов без знаков препинания и стоп-слов
filtered_tokens = [token.text for token in doc if is_token_valid(token)]

# Вывод списка токенов
print(filtered_tokens)
```

```python
word_frequencies = {}
for word in doc:
  if not word.is_punct:
    if not word.is_stop:
      if word.text not in word_frequencies.keys():
        word_frequencies[word.text] = 1
      else:
        word_frequencies[word.text] += 1
```

Начинаем ранжировать полученный список токенов

```python
from heapq import nlargest
select_length = int(len(word_frequencies))
summary = nlargest(select_length, word_frequencies, key = word_frequencies.get)
summary
```
