## Посмотрим на структуру файла

```python
import pandas as pd
import numpy as np
d_files = os.listdir(os.path.join(path,'idle'))
d_file = os.path.join(os.path.join(path,'idle'), d_files[0])
data_fr = pd.read_csv(d_file)
data_fr['activity'] = 0
#print(data_fr)
```

## Формируем датасет для активности idle, кодируемой числом 0

```python
d_files = os.listdir(os.path.join(path,'idle'))   # список файлов в папке
data_fr_0 = pd.DataFrame([])
for file_i in d_files:
  d_file = os.path.join(os.path.join(path,'idle'), file_i)
  d_fr = pd.read_csv(d_file)
  d_fr['activity'] = 0        # добавляем столбец - будущий y
  data_fr_0 = pd.concat([data_fr_0, d_fr], axis=0, ignore_index=True)  # соединяемм датасеты из файлов в один датасет
#print(data_fr_0.head(35))
#print(data_fr_0.tail(35))
if len(d_files) * 30 == data_fr_0.shape[0]:  ## проверка корректности датасета по длине
  print('Dataset IDLE convert correct')
else:
  print('Dataset IDLE conver UNcorrect - you should find a bag')
```

## формируем общий датасет

```python
dataset_target = pd.concat([data_fr_0, data_fr_1], axis=0, ignore_index=True)
dataset_target = pd.concat([dataset_target, data_fr_2], axis=0, ignore_index=True)
dataset_target = pd.concat([dataset_target, data_fr_3], axis=0, ignore_index=True)
print(dataset_target.shape)
print(dataset_target.columns)
```

## Готовим данные для обучения классификатора

```python
from sklearn.utils import shuffle
from sklearn import model_selection
df_shuff = shuffle(dataset_target, random_state=42)   # перемешаем датасет для валидности
X = np.array(df_shuff[['accelerometer_X', 'accelerometer_Y', 'accelerometer_Z']])  # матрица Х - пока пробуем без стандартизации данных
y = np.array(df_shuff['activity'])                                                 # вектор у
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, train_size=0.25) # разбили для оценки скора после обучения
```

# SVC классификатор - по сырым данным

```python
from sklearn.svm import SVC
clf = SVC(C=1.0, decision_function_shape='ovr') # one vs rest
clf.fit(X_train, y_train)   # обучили классификатор
print(clf.predict([[1.0, 1.0, 1.0]]))   # тестовое предсказание
print(clf.score(X_test, y_test))
```

имеем точность 0.89 - єто немного
проверим для стратегии - ovo



```python
clf = SVC(C=1.0, decision_function_shape='ovo') # one vs one
clf.fit(X_train, y_train)   # обучили классификатор
print(clf.predict([[1.0, 1.0, 1.0]]))   # тестовое предсказание
print(clf.score(X_test, y_test))
```

Видим что стратегия не особо улучшает точность модели. Попробуем с коэффициентом регуляризации

```python
clf = SVC(C=0.5, decision_function_shape='ovr') # one vs rest С=0.5
clf.fit(X_train, y_train)   # обучили классификатор
print(clf.predict([[1.0, 1.0, 1.0]]))   # тестовое предсказание
print(clf.score(X_test, y_test))
```

```python
clf = SVC(C=2.0, decision_function_shape='ovr') # one vs rest С=2.0
clf.fit(X_train, y_train)   # обучили классификатор
print(clf.predict([[1.0, 1.0, 1.0]]))   # тестовое предсказание
print(clf.score(X_test, y_test))
```

```python
clf = SVC(C=10.0, decision_function_shape='ovr') # one vs rest С=10.0
clf.fit(X_train, y_train)   # обучили классификатор
print(clf.predict([[1.0, 1.0, 1.0]]))   # тестовое предсказание
print(clf.score(X_test, y_test))
```

Итак - при сырых НАШИХ!!!  данных модель почти не меняет точность при смене стратегии, но увеличивает точность при увеличении коэффициента регуляризации

Пробуем все то же самое но модель - RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=4, random_state=0, bootstrap=True) # bootstrap data
clf.fit(X_train, y_train)   # обучили классификатор
print(clf.predict([[1.0, 1.0, 1.0]]))   # тестовое предсказание
print(clf.score(X_test, y_test))
```

```python
clf = RandomForestClassifier(max_depth=4, random_state=0, bootstrap=False) # not bootstrap data
clf.fit(X_train, y_train)   # обучили классификатор
print(clf.predict([[1.0, 1.0, 1.0]]))   # тестовое предсказание
print(clf.score(X_test, y_test))
```

```python
clf = RandomForestClassifier(max_depth=10, random_state=0, bootstrap=True) # bootstrap data
clf.fit(X_train, y_train)   # обучили классификатор
print(clf.predict([[1.0, 1.0, 1.0]]))   # тестовое предсказание
print(clf.score(X_test, y_test))
```

```python
clf = RandomForestClassifier(max_depth=20, random_state=0, bootstrap=True) # bootstrap data
clf.fit(X_train, y_train)   # обучили классификатор
print(clf.predict([[1.0, 1.0, 1.0]]))   # тестовое предсказание
print(clf.score(X_test, y_test))
```

Видим модель RandomForest работает намного быстрее чем SVC и точнее при оответствующей глубине построения дерева. В общем то именно это и написано в мануале, что SVC сильно замедляется если величина семплов превышает десятки тысяч, а у нас 193860 семплов.
